2. Data Acquisition

In this section, we load the data from Yahoo Finance using the yfinance library. 
+1

Python
import yfinance as yf
import pandas as pd

# Define the company and data access date 
ticker_symbol = "AAPL" # Apple Inc.
access_date = "2026-04-23" 

# Fetching financial data
company = yf.Ticker(ticker_symbol)
income_statement = company.financials
balance_sheet = company.balance_sheet

print(f"Data for {ticker_symbol} successfully loaded on {access_date}.")
3. Data Cleaning & Preparation

Raw data often contains missing values or is transposed. We need to clean it to ensure correct calculations. 

Python
# Transpose the data so that years are rows and accounts are columns
df_income = income_statement.T
df_balance = balance_sheet.T

# Selecting only relevant columns [cite: 315]
# Note: Specific row names might vary by ticker; cleaning ensures consistency.
df_clean = pd.DataFrame({
    'Net Income': df_income['Net Income'],
    'Total Revenue': df_income['Total Revenue'],
    'Total Equity': df_balance['Stockholders Equity']
})

# Removing any NaN values to avoid calculation errors
df_clean = df_clean.dropna()
print("Data cleaning complete.")
display(df_clean.head())
4. Substantive Analysis (Financial Ratios)

We will now calculate the Return on Equity (ROE) and Profit Margin.

# Calculating Profitability Ratios [cite: 14]
df_clean['ROE (%)'] = (df_clean['Net Income'] / df_clean['Total Equity']) * 100
df_clean['Profit Margin (%)'] = (df_clean['Net Income'] / df_clean['Total Revenue']) * 100

# Displaying the calculated metrics 
display(df_clean[['ROE (%)', 'Profit Margin (%)']])
5. Insights & Interpretation

Observation 1: If the ROE is increasing, it suggests the company is becoming more efficient at generating profit from shareholders' capital.

Observation 2: A stable Profit Margin indicates consistent cost management relative to sales.
